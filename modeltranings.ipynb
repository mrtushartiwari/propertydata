```python
import time
start = time.clock()
import pandas as pd
df = pd.read_pickle('cleanphase1')
print(start - time.clock())
```

```python
#df2 = .str.startswith('bhk')
templist =list()
for i in df.columns:
    if i.startswith('bhk') == True:
        templist.append(str(i))
        
for i in df.columns:
    if i.startswith('age') == True:
        templist.append(str(i))
for i in df.columns:
    if i.startswith('locality') == True:
        templist.append(str(i))
        
for i in df.columns:
    if i.startswith('neworold') == True:
        templist.append(str(i))
        
for i in df.columns:
    if i.startswith('additionalrooms') == True:
        templist.append(str(i))
        
for i in df.columns:
    if i.startswith('facing') == True:
        templist.append(str(i))

for i in df.columns:
    if i.startswith('overlooking') == True:
        templist.append(str(i))
        
for i in df.columns:
    if i.startswith('ownership') == True:
        templist.append(str(i))
        
for i in df.columns:
    if i.startswith('possesiondate') == True:
        templist.append(str(i))
templist 
```

```python
numeric_col = ['area','bathroom','balconies','Lift Available','Car Parking','Power Backup','24 X 7 Security',
          'Children\'s play area','Vaastu Compliant','Club House','Gymnasium','Swimming Pool','Sports Facility',
          'Indoor Games','Jogging Track','Maintenance Staff','Intercom','Golf Course','Cafeteria','Rain Water Harvesting',
          'Staff Quarter','Multipurpose Room','Landscaped Gardens','Shopping Mall','School','Hospital','ATM','AC',
           'Wardrobe','TV','Refrigerator','Sofa','Washing Machine','Wifi','Microwave','Dining Table','Gas connection',
           'BED','pricepersquare','roadfaceing','opensides','totalfloor',]
# 'bhk' is not a index
price_predictors = numeric_col
#numeric_col.extend(templist)

for i in templist:
    numeric_col.append(i)
    price_predictors.append(i)

numeric_col.append('price')
df2 = df[numeric_col]

#df2 = pd.concat([df, templist], axis=1)
```

## Building first pridictor

```python
y = df2.price
#price_predictors = ['area','bathroom','balconies','Lift Available','Car Parking','Power Backup','24 X 7 Security',
 #                   'Children\'s play area','Vaastu Compliant','Club House','Gymnasium','Swimming Pool','Sports Facility',
  #                 'Indoor Games','Jogging Track','Maintenance Staff','Intercom','Golf Course','Cafeteria','Rain Water Harvesting',
   #                'Staff Quarter','Multipurpose Room','Landscaped Gardens','Shopping Mall','School','Hospital','ATM','AC',
    #                'Wardrobe','TV','Refrigerator','Sofa','Washing Machine','Wifi','Microwave','Dining Table','Gas connection',
     #              'BED','pricepersquare','roadfaceing','opensides','totalfloor',]


X = df2[price_predictors]
X = X.drop('price', 1)
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model
house_model = DecisionTreeRegressor()

# Fit model
house_model.fit(X, y)
```

## DecisionTreeRegressor

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import explained_variance_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_squared_log_error
from sklearn.metrics import median_absolute_error
from sklearn.metrics import r2_score


# split data into training and validation data, for both predictors and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.


train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 5)
# Define model
house_model_DT = DecisionTreeRegressor()
# Fit model
house_model_DT.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions_DT = house_model.predict(val_X)
print("Mean Absolute Error",mean_absolute_error(val_y, val_predictions_DT))
print("Explained Variance Score",explained_variance_score(val_y, val_predictions_DT))
print("Mean Squared Error",mean_squared_error(val_y, val_predictions_DT))
print("Mean Squared Log Error",mean_squared_log_error(val_y, val_predictions_DT))
print("Median Absolute Error",median_absolute_error(val_y, val_predictions_DT))
print("r2 score",r2_score(val_y, val_predictions_DT))


```

```python
from sklearn.externals import joblib
joblib.dump(house_model_DT, 'DTmodel.pkl') 
```

## RandomForestRegressor

```python
model2time = time.clock()
from sklearn.ensemble import RandomForestRegressor

# Define model
house_model = RandomForestRegressor()

# Fit model
house_model.fit(X, y)
```

```python
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both predictors and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.

train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 1)
# Define model
house_model_RF = RandomForestRegressor()
# Fit model
house_model_RF.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions_RF = house_model.predict(val_X)
print("Mean Absolute Error",mean_absolute_error(val_y, val_predictions_RF))
print("Explained Variance Score",explained_variance_score(val_y, val_predictions_RF))
print("Mean Squared Error",mean_squared_error(val_y, val_predictions_RF))
print("Mean Squared Log Error",mean_squared_log_error(val_y, val_predictions_RF))
print("Median Absolute Error",median_absolute_error(val_y, val_predictions_RF))
print("r2 score",r2_score(val_y, val_predictions_RF))
```

```python
from sklearn.externals import joblib
joblib.dump(house_model_RF, 'RFmodel.pkl')
```
